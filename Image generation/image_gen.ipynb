```python
from keras.preprocessing.image import load_img, img_to_array,ImageDataGenerator
```

```python
img = load_img('/content/horse.jpg',target_size=(256,256,3))
```

```python
image_conv = img_to_array(img)
```

```python
image_conv.shape
```

```python
image_conv[0]
```

```python
x = image_conv.reshape((1,)+image_conv.shape)
x.shape
```

```python
dataGenerator = ImageDataGenerator(featurewise_center=True, width_shift_range=0.7, zoom_range=0.8,horizontal_flip=True, vertical_flip=True,rotation_range=45)
```

```python
i = 1
for batch in dataGenerator.flow(x,batch_size=1,save_to_dir='/content/horse_regen',save_format='png',save_prefix='horse'):
  i = i+1
  if(i>20):
    break
```
